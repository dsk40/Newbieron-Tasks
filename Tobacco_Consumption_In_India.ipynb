```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#The data is webscrapped from website-https://www.findeasy.in/tobacco-use-among-adults-in-india/ using Instant Data Scraper

df=pd.read_csv("Tobacco Consumption In India.csv")
```

```python
# Convert all float values in the DataFrame to strings
df = df.astype(str)
```

```python
df
```

```python
# Convert the "Women" and "Men" columns to strings
df['Women'] = df['Women'].astype(str)
df['Men'] = df['Men'].astype(str)
```

```python
# Remove rows with NaN values in any column
df = df.drop(36)
```

```python
# Drop the column you want to remove
column_to_drop = 'SL No'
df = df.drop(column_to_drop, axis=1)
```

```python
df
```

```python
# Create a bar plot for women's consumption
plt.figure(figsize=(12, 6))
plt.bar(df['States'], df['Women'], color='pink', label='Women')
plt.xlabel('States')
plt.ylabel('Tobacco Consumption in %')
plt.title('State-wise Tobacco Consumption - Women')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Create a bar plot for men's consumption
plt.figure(figsize=(12, 6))
plt.bar(df['States'], df['Men'], color='blue', label='Men')
plt.xlabel('States')
plt.ylabel('Tobacco Consumption in %')
plt.title('State-wise Tobacco Consumption - Men')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Calculate the total consumption and add a new column "Total"
df['Total'] = df['Women'] + df['Men']
```

```python
# Create a bar plot for total consumption
plt.figure(figsize=(12, 6))
plt.bar(df['States'], df['Total'], color='purple', label='Total')
plt.xlabel('States')
plt.ylabel('Tobacco Consumption in %')
plt.title('State-wise Total Tobacco Consumption')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Function to remove percentage sign and convert to float
def clean_percentage(value):
    if isinstance(value, float):
        return value
    return float(value.replace('%', ''))

# Clean the data in "Women" and "Men" columns by removing the percentage sign and converting to float
df['Women'] = df['Women'].apply(clean_percentage)
df['Men'] = df['Men'].apply(clean_percentage)
```

```python
# Create a stacked bar chart for men and women consumption
plt.figure(figsize=(10, 8))

# Plot men's consumption as a bar
plt.bar(df['States'], df['Men'], color='blue', label='Men')

# Plot women's consumption as a bar on top of men's consumption
plt.bar(df['States'], df['Women'], bottom=df['Men'], color='pink', label='Women')

plt.xlabel('States')
plt.ylabel('Tobacco Consumption in %')
plt.title('Comparison of Men and Women Tobacco Consumption')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Create a line chart for men and women consumption
plt.figure(figsize=(12, 10))

# Plot men's consumption as a line
plt.plot(df['States'], df['Men'], color='blue', label='Men')

# Plot women's consumption as a line
plt.plot(df['States'], df['Women'], color='pink', label='Women')

plt.xlabel('States')
plt.ylabel('Tobacco Consumption in %')
plt.title('Tobacco Consumption - Men vs. Women')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
