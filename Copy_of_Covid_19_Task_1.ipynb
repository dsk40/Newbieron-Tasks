```python
# Send a GET request to the website
url = "https://prsindia.org/covid-19/cases"
response = requests.get(url)
```

```python
# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Find the table containing the statewise data
table = soup.find("table")
```

```python
# Find all rows in the table
rows = table.find_all("tr")
```

```python
# Initialize lists to store the data
state_list = []
confirmed_list = []
active_list = []
recovered_list = []
```

```python
# Iterate over the rows and extract the data
for row in rows:
    # Get the cells in the row
    cells = row.find_all("td")

    # Extract the required data (e.g., state name and COVID-19 cases)
    if len(cells) >= 4:
        state = cells[0].text.strip()
        confirmed_cases = cells[1].text.strip()
        active_cases = cells[2].text.strip()
        recovered_cases = cells[3].text.strip()

        # Append the data to respective lists
        state_list.append(state)
        confirmed_list.append(confirmed_cases)
        active_list.append(active_cases)
        recovered_list.append(recovered_cases)
```

```python
# Create a DataFrame from the collected data
data = {
    "State": state_list,
    "Confirmed Cases": confirmed_list,
    "Active Cases": active_list,
    "Recovered Cases": recovered_list
}
df = pd.DataFrame(data)

```

```python
print(df)
```

```python
# Sort the DataFrame by confirmed cases in descending order
df = df.sort_values(by="Confirmed Cases", ascending=False)

# Set the figure size
plt.figure(figsize=(12, 6))

# Create the bar chart
plt.bar(df["State"], df["Confirmed Cases"])
plt.xticks(rotation=90)
plt.xlabel("State")
plt.ylabel("Confirmed Cases")
plt.title("COVID-19 Confirmed Cases Statewise")

# Display the chart
plt.show()
```

```python
# Send a GET request to the website
url = "https://www.mygov.in/corona-data/covid19-statewise-status/"
response = requests.get(url)
```

```python
# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")
```

```python
# Find the main container div containing the state-wise data
container = soup.find("div", class_="field field-name-field-covid-statewise-data field-type-field-collection field-label-above")
```

```python
# Find all rows in the table
rows = container.find_all("div", class_="field-items")
```

```python
# Initialize lists to store the data
state_list = []
confirmed_list = []
active_list = []
recovered_list = []
deaths_list = []
```

```python
# Iterate over the rows and extract the data
for row in rows:
    # Get the cells in the row
    cells = row.find_all("div", class_="field-item")

    # Extract the required data (e.g., state name, confirmed cases, active cases, recovered cases, deaths)
    if len(cells) >= 5:
        state = cells[0].text.strip()
        confirmed_cases = cells[1].text.strip()
        active_cases = cells[2].text.strip()
        recovered_cases = cells[3].text.strip()
        deaths = cells[4].text.strip()

        # Append the data to respective lists
        state_list.append(state)
        confirmed_list.append(confirmed_cases)
        active_list.append(active_cases)
        recovered_list.append(recovered_cases)
        deaths_list.append(deaths)

```

```python
# Create a DataFrame from the collected data
data = {
    "State": state_list,
    "Confirmed Cases": confirmed_list,
    "Active Cases": active_list,
    "Recovered Cases": recovered_list,
    "Deaths": deaths_list
}
```
