```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/IDSCountry.csv")
```

```python
# Summary Statistics and Data Information
print(df.describe())
```

```python
# Perform data cleaning (if required)
# Example: Drop rows with missing values in specific columns
columns_to_clean = ['Income Group', 'Region', 'Currency Unit']
df_cleaned = df.dropna(subset=columns_to_clean)
```

```python
# Basic EDA: Visualizing the distribution of a numeric column
numeric_column = 'Latest population census'
plt.figure(figsize=(8, 6))
plt.hist(df_cleaned[numeric_column], bins=20, edgecolor='black')
plt.xlabel(numeric_column)
plt.xticks(rotation=45)
plt.ylabel('Frequency')
plt.title(f'Distribution of {numeric_column}')
plt.show()
```

```python
# Check the frequency distribution of the 'Lending category' column
lending_category_counts = df['Lending category'].value_counts()

# Create a bar plot to visualize the distribution
plt.figure(figsize=(10, 6))
plt.bar(lending_category_counts.index, lending_category_counts.values)
plt.xlabel('Lending Category')
plt.ylabel('Count')
plt.title('Frequency Distribution of Lending Category')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
income_group_counts = df['Income Group'].value_counts()

# Create a bar plot to visualize the distribution
plt.figure(figsize=(10, 6))
plt.bar(income_group_counts.index, income_group_counts.values)
plt.xlabel('Income Group')
plt.ylabel('Count')
plt.title('Frequency Distribution of Income Groups')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
region_counts = df['Region'].value_counts()

# Define a list of colors for each region (You can customize this list based on your preferences)
colors = ['blue', 'green', 'orange', 'purple', 'red', 'brown', 'pink']

# Create a horizontal bar chart with different colors for each bar
plt.figure(figsize=(10, 6))
plt.barh(region_counts.index, region_counts.values, color=colors)
plt.xlabel('Count')
plt.ylabel('Region')
plt.title('Frequency Distribution of Countries in Each Region')
plt.tight_layout()
plt.show()
```

```python
#Correlation Heatmap
numeric_columns = df.select_dtypes(include='number').columns
correlation_matrix = df[numeric_columns].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Box Plot for Numeric Columns
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[numeric_columns])
plt.xticks(rotation=90)
plt.title('Box Plot for Numeric Columns')
plt.show()
```

```python
# Pair Plot
sns.pairplot(df[numeric_columns])
plt.suptitle('Pair Plot for Numeric Columns', y=1.02)
plt.show()
```
